W6A2

Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu tới vị
trí i), giữ nguyên thứ tự

```python
lst = list(map(int, input().split()))

def total(lst):
    lst1 = []
    current_sum = 0
    for x in lst:
        current_sum += x      # Cộng dồn giá trị hiện tại vào tổng
        lst1.append(current_sum)
    return lst1

print(total(lst))
```

Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp
xếp.)


```python
t = tuple(map(int, input().split()))
k = int(input())

n = len(t)
k = k % n  # tránh k > n

res = t[k:] + t[:k]
print(res)
```

W6A4

Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key lặp
lại, thêm vào list của key đó

```python
s = input().split()

d = {}

for i in s:
    key, value = i.split(':')
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)
print(d) 
```

W6A5

Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 'zeros':
...} (đếm số >0, <0, =0

```python
s = map(int, input().split())

countp = 0
countn = 0
countz = 0
for i in s:
    if i > 0:
        countp += 1
    if i < 0:
        countn += 1
    if i == 0:
        countz += 1
d = {'positives': countp, 'negatives': countn, 'zeros': countz}
print(d)


```
