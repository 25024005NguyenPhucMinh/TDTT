nguyên, phần dư và kết quả chia thực của phép chia a cho b (lấy đến 2 chữ số phần thập
phân) theo định dạng đúng của output

W2A4 [AverageOfSix]
Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng
6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6
đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được
phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

```python
a1, b1, c1, a2, b2, a3 = map(int, input().split())
Tb = round(((a1 + b1 + c1) + (a2 + b2) * 2 + a3 * 3) / 10, 1) 
print (Tb)
```

Nhập vào họ tên chủ hộ và chỉ số điện kế tháng trước và chỉ số điện kế tháng này,
tính tiền điện tháng này cho hộ biết:
- Từ 0 – 50kWh: 1.984 đồng/kWh
- Bậc 2: Từ 51 – 100 kWh: 2.050 đồng/kWh
- Bậc 3: Từ 101 – 200 kWh: 2.380 đồng/kWh
- Bậc 4: Từ 201 – 300 kWh: 2.998 đồng/kWh
- Bậc 5: Từ 301 – 400 kWh: 3.350 đồng/kWh
- Bậc 6: Từ 401 trở đi kWh: 3.460 đồng/kWh 

```python
X = input('Ten chu ho:')
a = float(input('chi so thang truoc:'))
b = float(input('chi so thang nay:'))
def compute(a, b):
    d = b - a
    if 0 < d <= 50:
        return d * 1984
    elif d <= 100:
        return 50 * 1984 + (d - 50) * 2050
    elif d <= 200:
        return 50 * 1984 + 50 * 2050 + (d - 100) * 2380
    elif d <= 300:
        return 50 * 1984 + 50 * 2050 + 100 * 2380 + (d - 200) * 2998
    elif d <= 400:
        return 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + (d - 300) * 3350
    else:
        return 50 * 1984 + 50 * 2050 + 100 * 2380 + 100 * 2998 + 100 * 3350 + (d - 400) * 3460

m = compute(a, b)
m = m * 1.08
print(f"Ho va ten: {X} \nTien phai tra la: {m}")
```

Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”.

```python
a = float(input())
b = float(input())
c = float(input())
def compute(a, b, c):
    d = (a + b + c) / 2
    if a + b > c and a + c > b and b + c > a:
        return (d * (d - a) * (d - b) * (d - c)) ** 0.5
    else:
        print("Khong phai la 3 canh cua 1 tam giac")
print (f"{compute(a, b, c):.1f}")
```

Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt.
