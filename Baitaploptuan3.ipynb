W3A3

Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
a = int(input())
if 0 < a and a & (a - 1) == 0:
    print("la luy thua cua 2")
else:   
    print("khong la luy thua cua 2")
```

[Even and Odd]

Nhập vào một số nguyên dương x, bạn hãy kiểm tra xem x là số chẵn hay lẻ. Nếu x là số chẵn, in ra "Even". Nếu x
là số lẻ, in ra "Odd". 


```python
a = int(input())
if a % 2 == 0:
    print("Even")
elif a % 2 == 1:
    print("Odd")
```

W3A7

[Check negative numbers]

Nhập vào hai số nguyên a, b. Nếu cả hai số a, b đều âm thì in ra "Yes", ngược lại in ra "No".

```python
a, b = map(int, input().split())
if a < 0 and b < 0:
    print("Yes")
else:
    print("No")
```

[Longer string]

Nhập vào hai chuỗi a, b gồm các chữ cái thường. In ra "True" nếu chuỗi a dài hơn chuỗi b. Ngược lại in ra "False".

```python
a, b = input().split()
if len(a) > len(b):
    print("True")
else:    
    print("False")
```
