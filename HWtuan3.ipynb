Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểm tra xem nó có thể là độ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra màn hình.

```python
a, b ,c = map(float, input().split()) 
def check(a, b ,c):   
  if a + b > c and a + c > b and b + c > a and not (a == b == c) and not (a == b or a == c or b == c) and not (a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2):    
    print ('Tam giac thuong')     
  elif a == b == c:     
    print ('Tam giac deu')     
  elif (a == b or a == c or b == c):      
    print ('Tam giac can')     
  elif (a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2):      
    print ('Tam giac vuong')     
  else:       
    print ('Khong phai tam giac') 
check(a, b, c)
def compute(a, b, c):     
    p = (a + b + c) / 2     
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5     
    return s 
print(compute(a, b, c))
print(a + b + c)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax^2 + bx + c = 0)

```python
a, b, c = map(float, input().split())
def compute(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return 'Phuong trinh co vo so nghiem'
            else:
                return 'Phuong trinh vo nghiem'
        else:
            x = - c / b
            return f'Phuong trinh co mot nghiem: {x}'
    delta = b ** 2 - 4 * a * c
    if delta < 0:
        return 'Phuong trinh vo nghiem'
    elif delta == 0:
        x = - b / (2 * a)
        return f'Phuong trinh co nghiem kep: {x}'
    else:
        x1 = (- b + delta ** 0.5) / (2 * a)
        x2 = (- b - delta ** 0.5) / (2 * a)
        return f'Phuong trinh co hai nghiem phan biet: {x1} va {x2}'
print(compute(a, b, c))
```

Bài 8: Giải hệ phương trình tuyến tính:

{ax + by = m

{cx + dy = n


```python
a, b, c, d, m, n = map(float, input().split())

def compute(a, b, c, d, m, n):
    delta = a * d - b * c
    delta1 = m * d - b * n
    delta2 = a * n - m * c

    if delta != 0:
        x = delta1 / delta
        y = delta2 / delta
        return f'He phuong trinh co nghiem duy nhat: x = {x}, y = {y}'
    else:
        if delta1 == 0 and delta2 == 0:
            return 'He phuong trinh co vo so nghiem'
        else:
            return 'He phuong trinh vo nghiem'

print(compute(a, b, c, d, m, n))
```

Bài 10:

 Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
x0, y0 = map(float, input("Nhap tam (x0 y0): ").split())
R = float(input("Nhap ban kinh: "))
x, y = map(float, input("Nhap diem A (x y): ").split())
d = ((x - x0) ** 2 + (y - y0) ** 2) ** 0.5
if d == R: 
    print("Diem A thuoc duong tron")
else: 
    print("Diem A khong thuoc duong tron")
```
