Yêu cầu. Cước phí Taxi Thông Minh.
Viết một hàm duy nhất smart_taxi_fare(car_type, start_hour, km, wait_minutes) tính và trả về total (tổng tiền phải trả,
kiểu số nguyên, đơn vị VND).
Các quy tắc cần cài đặt:
1. Giá mở cửa (0–0.8 km):
- Xe 4 chỗ: 11,000 VND
- Xe 7 chỗ: 12,000 VND
2. Giá theo km sau 0.8 km:
- Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
- Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km
3. Phí chờ: 5 phút đầu miễn phí; sau đó, mỗi 5 phút (làm tròn lên theo từng block 5 phút) tính 2,000 VND.
4. Phụ thu ban đêm: nếu start_hour thuộc [22,23,0,1,2,3,4,5], cộng +20% trên tiền tạm tính (mở cửa + quãng đường +
chờ).
5. Điều kiện đặc biệt:
- Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
- Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu). 

```python
car_type, start_hour, km, wait_minutes = map(float, input().split())
import math
def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    distance_fare = 0
    if car_type == 4:
        if 0 <= km <= 0.8:
            distance_fare += 11000
        elif 0.8 < km <= 20:
            distance_fare = 11000 + 13000 * (km - 0.8)
        elif 20 < km:
            distance_fare = 11000 + 13000 * (20 - 0.8) + 10500 * (km - 20)
    elif car_type == 7:
        if 0 <= km <= 0.8:
            distance_fare += 12000
        elif 0.8 < km <= 20:
            distance_fare = 12000 + 14000 * (km - 0.8)
        elif 20 < km:
            distance_fare = 12000 + 14000 * (20 - 0.8) + 12000 * (km - 20)
    wait_fare = 0
    if wait_minutes > 5:
        wait = wait_minutes - 5
        block = math.ceil(wait / 5)
        wait_fare = block * 2000
    subtotal = distance_fare + wait_fare
    is_night = start_hour >= 22 or start_hour <= 5
    if is_night:
        subtotal = subtotal * 1.2 
    if km > 30 and not is_night:
        subtotal = subtotal * 0.9
    total = int(subtotal)
    if total < 50000 and km > 5:
        total = 50000
    return total
print(smart_taxi_fare(car_type, start_hour, km, wait_minutes))
```

Yêu cầu: Cho số nguyên x, trả về True nếu tồn tại số nguyên n ≥ 1 sao cho x = 1 + 2 + ... + n, ngược lại trả về False.
Khuyến khích dùng vòng lặp đơn giản để cộng dồn tổng

```python
x = int(input())

def sum(x):
  s = 0
  for i in range(1, x):
    s += i
    if s == x:
      return True
    if s > x:
      return False
  return False
print(sum(x))
```

Yêu cầu: Cho các hệ số thực a, b, c, trả về tất cả các nghiệm thực của phương trình bậc hai axx + bx + c = 0 trong một danh
sách được sắp xếp tăng dần.
Quy tắc:
- Nếu a == 0 và b != 0: trả về nghiệm duy nhất [-c / b].
- Nếu a == 0 và b == 0: trả về [] (không có nghiệm hữu hạn duy nhất nếu c == 0, hoặc vô nghiệm nếu c != 0).
- Với a != 0, dùng biệt thức d = bb - 4ac:
- d < 0 → []
- d == 0 → [ -b / (2a) ]
- d > 0 → hai nghiệm [r1, r2] với r1 <= r2.


```python
a, b, c = map(int, input().split())
def pt(a, b, c):
    if a == 0:
        if b == 0:
            if c == 0:
                return "Infinite solutions"
            else:
                return "No solution"
        else:
            return f"One solution: x = {-c/b}"
    else:
        d = b ** 2 - 4 * a * c
        if d < 0:
            return "No real solutions"
        elif d == 0:
            return f"One real solution: x = {-b/(2*a)}"
        else:
            x1 = (-b + d**0.5) / (2*a)
            x2 = (-b - d**0.5) / (2*a)
            return f"Two real solutions: x1 = {x1}, x2 = {x2}"
print(pt(a, b, c))
```

Yêu cầu: Cho hai ngày sinh dưới dạng (day, month, year) của người 1 và người 2, trả về 1 nếu người 1 lớn tuổi hơn (ngày
sinh sớm hơn), 2 nếu người 2 lớn tuổi hơn, và 0 nếu bằng nhau.
Ví dụ: 27 2 2007 so với 28 2 2007 → trả về 1 vì ngày 27 sớm hơn nên người 1 lớn tuổi hơn. 

```python
a1, b1, c1, a2, b2, c2 = map(int, input().split())
p1 = (a1, b1, c1)
p2 = (a2, b2, c2)
def older(p1, p2):
    if p1[2] < p2[2]:
        return "1"
    elif p1[2] > p2[2]:
        return "2"
    elif p1[2] == p2[2]:
        if p1[1] < p2[1]:
            return "1"
        elif p1[1] > p2[1]:
            return "2"
        elif p1[1] == p2[1]:
            if p1[0] < p2[0]:
                return "1"
            elif p1[0] > p2[0]:
                return "2"
            else:
                return "0"
print(older(p1, p2))
```

Yêu cầu: Cho số nguyên n, trả về:
- 5n + 1 nếu n mod 3 == 1
- 6n + 2 nếu n mod 3 == 2
- 7n + 3 nếu n mod 3 == 0 


```python
n = int(input())
def mod(n):
    if n % 3 == 1:
        return 5 * n+ 1
    elif n % 3 == 2:
        return 6 * n + 2
    elif n % 3 == 0:
        return 7 * n + 3
print(mod(n))
    
```

Yêu cầu: Cho cân nặng (kg) và chiều cao (m) của hai người, hãy trả về người có BMI cao hơn.
Trả về 1 nếu người 1 có BMI cao hơn, 2 nếu người 2 có BMI cao hơn, và 0 nếu bằng nhau.
Nhắc lại: BMI = weight / (height height). 


```python
a1, b1, a2, b2 = map(float, input().split())
p1 = (a1, b1)
p2 = (a2, b2)
def BMI(x, y):
    bmi = x / (y ** 2)
    return bmi
def compare_bmi(p1, p2):
    bmi1 = BMI(p1[0], p1[1])
    bmi2 = BMI(p2[0], p2[1])
    if bmi1 < bmi2:
        return "2"
    elif bmi1 > bmi2:
        return "1"
    else:
        return "0"
print(compare_bmi(p1, p2))
```

```python
n = int(input())
def sum_of_squares(n):
    s = 0
    for i in range(1, n+1):
        s += i ** 2
    return s
print(sum_of_squares(n))  
```
