Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x

```python
x = int(input())
X = str(x)
s = 0
def compute(x, X, s):
    for i in range(len(X)):
        s += int(X[i]) ** len(X)
    if s == x: return 'True'
    else: return "False"
print(compute(x, X, s))        
```

W5A17

Viết hàm kiêểm tra một số có phải số chính phương hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x


```python
n = int(input())
def bool(n):
    for i in range(n):
        if i * i == n:
            return True
    return False
print(bool(n))
```

VIết hàm kiểm tra một số có phải số đối xứng không, xử lý trên dữ liệu kiểu số nguyên

```python
x = int(input())

def palidrome(x):
    return str(x) == str(x)[::-1]
print(palidrome(x))
```

VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên.


```python
x = int(input())
def magic(x):
    while x > 9:
        s = 0
        for i in str(x):
            s += int(i)
        x = s
    return x == 1
print(magic(x))
```

W5A20 

Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'

```python
s = input()
def detect(s):
    try: 
        if '.' in s:
            float(s)
            return 'float'
        else: 
            int(s)
            return 'int'
    except:
        return 'neither'
print(detect(s))
```

Viết hàm is_one_product_of_others(a, b, c) trả về True nếu một trong ba số bằng tích của hai số còn lại

```python
a, b, c = map(float, input().split())
def check(a, b, c):
    if a * b == c or a * c == b or b * c == a:
        return True
    return False
print(check(a, b, c))
```
